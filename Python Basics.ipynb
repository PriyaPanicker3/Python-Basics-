2.   What is an interpreter in Python?

  An interpreter in Python is a program that reads and executes Python code line by line. It converts high-level Python code into machine-readable instructions at runtime, allowing you to run your code without compiling it first.

5. What is mutability in Python

  In Python, mutability refers to an object's ability to be changed after it is created. If an object is mutable, its contents can be modified without changing its identity. Common mutable types include lists, dictionaries, and sets.

6. Why are lists mutable, but tuples are immutable ?

  LLists are mutable because they are made to store data that can change. You can add, remove, or change items in a list.

  Tuples are immutable because they are meant to store data that should not change. Once created, you cannot change the items in a tuple.

7. What is the difference between “==” and “is” operators in Python?

In Python, == is the equality operator that checks whether two variables have the same value. It compares the contents of the variables.

The is operator is the identity operator that checks whether two variables refer to the same object in memory. It compares the object identities, not their values.

9. What is type casting in Python?

Type casting in Python means converting one data type into another.

10. What is the difference between implicit and explicit type casting?
  
  In Python, implicit type casting happens automatically when Python changes one data type to another during a calculation. For example, if you add an integer and a float, Python changes the integer to a float by itself. Explicit type casting is when you manually change the type using functions like int(), float(), or str(). For example, you can turn the string "10" into the number 10 using int("10"). Implicit casting is done by Python, while explicit casting is done by the programmer.

13. What is the difference between for and while loops ?

- for loop is used when you know in advance how many times you want to repeat a block of code (e.g., looping through a list or a range).

- while loop is used when you don’t know exactly how many times to loop — it runs as long as a condition is true.

- for loops work with sequences, while while loops depend on a condition.

- In while loops, you must manually update variables to avoid infinite loops.

14. Describe a scenario where a while loop is more suitable than a for loop ?

A while loop is more suitable in scenarios where the number of iterations is not known in advance and depends on a condition that changes dynamically during execution.
