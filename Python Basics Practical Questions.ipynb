```python
# 4. Write a program that checks if a given word is a Python keyword.

import keyword

word = input("Enter a word: ")

if keyword.iskeyword(word):
    print(word, "is a Python keyword.")
else:
    print(word, "is not a Python keyword.")
```

```python
# 5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.

my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

# Changing an element in the list
print("Original list:", my_list)
my_list[1] = 99
print("List after change:", my_list)

# Trying to change an element in the tuple
print("\nOriginal tuple:", my_tuple)
try:
    my_tuple[1] = 99
except TypeError as e:
    print("Error when trying to change tuple:", e)



```

```python
# 6. Write a function to demonstrate the behavior of mutable and immutable

def change_values(a, b):
    a = a + 5          # Try to change number (immutable)
    b.append(10)       # Try to change list (mutable)
    print("Inside function:")
    print("a:", a)
    print("b:", b)

x = 5
y = [1, 2, 3]

print("Before function call:")
print("x:", x)
print("y:", y)

change_values(x, y)

print("\nAfter function call:")
print("x:", x)  # unchanged
print("y:", y)  # changed


```

```python
# 7. Write a program that performs basic arithmetic operations on two user-input numbers.

# Get input from user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Perform arithmetic operations
print("\nResults:")
print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)

# Avoid division by zero
if num2 != 0:
    print("Division:", num1 / num2)
else:
    print("Division: Cannot divide by zero")
```

```python
# 13. Write a Python program to find the sum of all even numbers between 1 and 50

total = 0

for i in range(2, 51, 2):
    total += i

print("Sum of even numbers from 1 to 50:", total)
```

```python
# 14. Write a program to reverse a string using a while loop.

s = input("Enter a string: ")
i = len(s) - 1
rev = ""

while i >= 0:
    rev += s[i]
    i -= 1

print("Reversed:", rev)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a number: "))
factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is", factorial)
```
